DATA CLEANING AND EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

print("Libraries imported successfully! ✓")
print(f"Pandas version: {pd.__version__}")
print(f"NumPy version: {np.__version__}")
```

```python
# ════════════════════════════════════════════
# CELL 2: LOAD DATA
# ════════════════════════════════════════════
print("="*60)
print("LOADING DATA")
print("="*60)

# Load the file (it's in the current directory)
df = pd.read_csv('household_power_consumption.txt',
                 sep=';',
                 parse_dates={'DateTime': ['Date', 'Time']},
                 infer_datetime_format=True,
                 na_values=['?'],
                 low_memory=False)

print("\n✓ DATA LOADED SUCCESSFULLY!")
print("="*60)

print(f"\nDataset Shape: {df.shape}")
print(f"  • Rows: {df.shape[0]:,}")
print(f"  • Columns: {df.shape[1]}")

print(f"\nDate Range:")
print(f"  • Start: {df['DateTime'].min()}")
print(f"  • End: {df['DateTime'].max()}")
print(f"  • Duration: {(df['DateTime'].max() - df['DateTime'].min()).days:,} days")

print(f"\nColumn Names:")
for i, col in enumerate(df.columns, 1):
    print(f"  {i}. {col}")

print("\n" + "-"*60)
print("First 5 rows:")
print("-"*60)
df.head()
```

```python
# ════════════════════════════════════════════
# CELL 3: DATA EXPLORATION
# ════════════════════════════════════════════
print("="*60)
print("DATA EXPLORATION")
print("="*60)

print("\nDataset Info:")
df.info()

print("\n" + "-"*60)
print("Statistical Summary:")
print("-"*60)
df.describe()
```

```python
# ════════════════════════════════════════════
# CELL 4: MISSING VALUES ANALYSIS
# ════════════════════════════════════════════
print("="*60)
print("MISSING VALUES ANALYSIS")
print("="*60)

missing_count = df.isnull().sum()
missing_pct = (missing_count / len(df)) * 100

missing_df = pd.DataFrame({
    'Column': missing_count.index,
    'Missing_Count': missing_count.values,
    'Missing_Percentage': missing_pct.values
})

missing_df = missing_df[missing_df['Missing_Count'] > 0].sort_values('Missing_Count', ascending=False)

if len(missing_df) > 0:
    print("\n⚠️  Columns with Missing Values:")
    print(missing_df.to_string(index=False))

    # Visualize
    plt.figure(figsize=(10, 6))
    plt.barh(missing_df['Column'], missing_df['Missing_Percentage'], color='coral', edgecolor='black')
    plt.xlabel('Missing Percentage (%)', fontsize=12)
    plt.ylabel('Column', fontsize=12)
    plt.title('Missing Values by Column', fontsize=14, fontweight='bold')
    plt.grid(axis='x', alpha=0.3)
    plt.tight_layout()
    plt.show()

    total_missing = missing_count.sum()
    total_cells = df.shape[0] * df.shape[1]
    print(f"\nTotal missing values: {total_missing:,}")
    print(f"Percentage of entire dataset: {(total_missing / total_cells * 100):.2f}%")
else:
    print("\n✓ No missing values found!")
```

```python
# ════════════════════════════════════════════
# CELL 5: DATA CLEANING
# ════════════════════════════════════════════
print("="*60)
print("DATA CLEANING")
print("="*60)

print(f"\nOriginal dataset:")
print(f"  • Rows: {df.shape[0]:,}")
print(f"  • Columns: {df.shape[1]}")

# Remove rows with missing values
df_cleaned = df.dropna()

print(f"\nCleaned dataset:")
print(f"  • Rows: {df_cleaned.shape[0]:,}")
print(f"  • Columns: {df_cleaned.shape[1]}")

rows_removed = df.shape[0] - df_cleaned.shape[0]
pct_removed = (rows_removed / df.shape[0]) * 100

print(f"\nData removed:")
print(f"  • Rows removed: {rows_removed:,}")
print(f"  • Percentage removed: {pct_removed:.2f}%")
print(f"  • Rows retained: {100-pct_removed:.2f}%")

# Verify cleaning
remaining_missing = df_cleaned.isnull().sum().sum()
print(f"\nRemaining missing values: {remaining_missing}")

if remaining_missing == 0:
    print("✓ All missing values successfully removed!")
else:
    print("⚠️  Warning: Some missing values still present!")

print("\n" + "="*60)
print("✓ DATA CLEANING COMPLETE")
print("="*60)
```

```python
# ════════════════════════════════════════════
# CELL 6: TIME SERIES VISUALIZATION
# ════════════════════════════════════════════
print("="*60)
print("TIME SERIES VISUALIZATION")
print("="*60)

fig, axes = plt.subplots(3, 1, figsize=(16, 12))

# Plot 1: Global Active Power
axes[0].plot(df_cleaned['DateTime'], df_cleaned['Global_active_power'],
             linewidth=0.5, color='steelblue', alpha=0.7)
axes[0].set_xlabel('Date', fontsize=11)
axes[0].set_ylabel('Power (kW)', fontsize=11)
axes[0].set_title('Global Active Power Over Time', fontsize=13, fontweight='bold', pad=10)
axes[0].grid(True, alpha=0.3)
axes[0].tick_params(axis='x', rotation=45)

# Plot 2: Voltage
axes[1].plot(df_cleaned['DateTime'], df_cleaned['Voltage'],
             linewidth=0.5, color='green', alpha=0.7)
axes[1].set_xlabel('Date', fontsize=11)
axes[1].set_ylabel('Voltage (V)', fontsize=11)
axes[1].set_title('Voltage Over Time', fontsize=13, fontweight='bold', pad=10)
axes[1].grid(True, alpha=0.3)
axes[1].tick_params(axis='x', rotation=45)

# Plot 3: Global Intensity
axes[2].plot(df_cleaned['DateTime'], df_cleaned['Global_intensity'],
             linewidth=0.5, color='red', alpha=0.7)
axes[2].set_xlabel('Date', fontsize=11)
axes[2].set_ylabel('Intensity (A)', fontsize=11)
axes[2].set_title('Global Intensity Over Time', fontsize=13, fontweight='bold', pad=10)
axes[2].grid(True, alpha=0.3)
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

print("\n✓ Time series plots generated!")
```

```python
# ════════════════════════════════════════════
# CELL 7: DISTRIBUTION ANALYSIS
# ════════════════════════════════════════════
print("="*60)
print("DISTRIBUTION ANALYSIS")
print("="*60)

numeric_cols = ['Global_active_power', 'Global_reactive_power', 'Voltage',
                'Global_intensity', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']

fig, axes = plt.subplots(3, 3, figsize=(16, 12))
axes = axes.ravel()

for idx, col in enumerate(numeric_cols):
    axes[idx].hist(df_cleaned[col], bins=50, color='skyblue',
                   edgecolor='black', alpha=0.7)
    axes[idx].set_xlabel(col, fontsize=10)
    axes[idx].set_ylabel('Frequency', fontsize=10)
    axes[idx].set_title(f'Distribution of {col}', fontsize=11, fontweight='bold')
    axes[idx].grid(True, alpha=0.3, axis='y')

    # Add statistics
    mean_val = df_cleaned[col].mean()
    median_val = df_cleaned[col].median()
    axes[idx].axvline(mean_val, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_val:.2f}')
    axes[idx].axvline(median_val, color='green', linestyle='--', linewidth=2, label=f'Median: {median_val:.2f}')
    axes[idx].legend(fontsize=8)

# Hide unused subplots
for idx in range(len(numeric_cols), len(axes)):
    axes[idx].axis('off')

plt.tight_layout()
plt.show()

print("\n✓ Distribution histograms generated!")
```

```python
# ════════════════════════════════════════════
# CELL 8: BOX PLOTS
# ════════════════════════════════════════════
print("="*60)
print("BOX PLOT ANALYSIS")
print("="*60)

fig, axes = plt.subplots(3, 3, figsize=(16, 12))
axes = axes.ravel()

for idx, col in enumerate(numeric_cols):
    bp = axes[idx].boxplot(df_cleaned[col], patch_artist=True)

    # Customize box plot
    for patch in bp['boxes']:
        patch.set_facecolor('lightblue')
    for whisker in bp['whiskers']:
        whisker.set(color='blue', linewidth=1.5)
    for cap in bp['caps']:
        cap.set(color='blue', linewidth=1.5)
    for median in bp['medians']:
        median.set(color='red', linewidth=2)

    axes[idx].set_ylabel(col, fontsize=10)
    axes[idx].set_title(f'Box Plot: {col}', fontsize=11, fontweight='bold')
    axes[idx].grid(True, alpha=0.3, axis='y')

# Hide unused subplots
for idx in range(len(numeric_cols), len(axes)):
    axes[idx].axis('off')

plt.tight_layout()
plt.show()

print("\n✓ Box plots generated!")
```

```python
# ════════════════════════════════════════════
# CELL 9: COVARIANCE ANALYSIS
# ════════════════════════════════════════════
print("="*60)
print("COVARIANCE ANALYSIS")
print("="*60)

numeric_cols = ['Global_active_power', 'Global_reactive_power', 'Voltage',
                'Global_intensity', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']

# Calculate covariance matrix
cov_matrix = df_cleaned[numeric_cols].cov()

print("\nCovariance Matrix:")
print(cov_matrix)

# Visualize covariance matrix
plt.figure(figsize=(11, 9))
sns.heatmap(cov_matrix, annot=True, fmt='.2e', cmap='coolwarm',
            square=True, cbar_kws={"shrink": 0.8}, linewidths=1,
            annot_kws={'fontsize': 9})
plt.title('Covariance Matrix of Power Consumption Variables',
          fontsize=14, fontweight='bold', pad=15)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

print("\n✓ Covariance matrix generated!")
```

```python
# ════════════════════════════════════════════
# CELL 10: CORRELATION ANALYSIS
# ════════════════════════════════════════════
print("="*60)
print("CORRELATION ANALYSIS")
print("="*60)

# Calculate correlation matrix
corr_matrix = df_cleaned[numeric_cols].corr()

print("\nCorrelation Matrix:")
print(corr_matrix)

# Visualize correlation matrix
plt.figure(figsize=(11, 9))
sns.heatmap(corr_matrix, annot=True, fmt='.3f', cmap='viridis',
            square=True, vmin=-1, vmax=1, center=0,
            cbar_kws={"shrink": 0.8}, linewidths=1,
            annot_kws={'fontsize': 10})
plt.title('Correlation Matrix of Power Consumption Variables',
          fontsize=14, fontweight='bold', pad=15)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

print("\n✓ Correlation matrix generated!")

# Print strongest correlations
print("\n" + "-"*60)
print("Strongest Correlations with Global Active Power:")
print("-"*60)
correlations = corr_matrix['Global_active_power'].drop('Global_active_power').sort_values(ascending=False)
for var, corr_val in correlations.items():
    if abs(corr_val) > 0.5:
        strength = "Strong" if abs(corr_val) > 0.7 else "Moderate"
        direction = "positive" if corr_val > 0 else "negative"
        print(f"  • {var}: {corr_val:.3f} ({strength} {direction})")
```

```python
# ════════════════════════════════════════════
# CELL 11: HOURLY PATTERN ANALYSIS
# ════════════════════════════════════════════
print("="*60)
print("HOURLY PATTERN ANALYSIS")
print("="*60)

# Extract hour from DateTime
df_cleaned['Hour'] = df_cleaned['DateTime'].dt.hour

# Calculate hourly averages
hourly_avg = df_cleaned.groupby('Hour')['Global_active_power'].agg(['mean', 'std', 'min', 'max'])

print("\nHourly Statistics:")
print(hourly_avg)

# Plot hourly pattern
fig, axes = plt.subplots(2, 1, figsize=(14, 10))

# Plot 1: Average power by hour
axes[0].bar(hourly_avg.index, hourly_avg['mean'], color='steelblue',
            edgecolor='black', alpha=0.7)
axes[0].errorbar(hourly_avg.index, hourly_avg['mean'], yerr=hourly_avg['std'],
                 fmt='none', color='red', capsize=3, alpha=0.5, label='Std Dev')
axes[0].set_xlabel('Hour of Day', fontsize=12)
axes[0].set_ylabel('Average Power (kW)', fontsize=12)
axes[0].set_title('Average Power Consumption by Hour of Day', fontsize=13, fontweight='bold')
axes[0].set_xticks(range(0, 24))
axes[0].grid(True, alpha=0.3, axis='y')
axes[0].legend()

# Plot 2: Min and Max by hour
axes[1].plot(hourly_avg.index, hourly_avg['max'], marker='o',
             label='Max', color='red', linewidth=2)
axes[1].plot(hourly_avg.index, hourly_avg['mean'], marker='s',
             label='Mean', color='blue', linewidth=2)
axes[1].plot(hourly_avg.index, hourly_avg['min'], marker='^',
             label='Min', color='green', linewidth=2)
axes[1].set_xlabel('Hour of Day', fontsize=12)
axes[1].set_ylabel('Power (kW)', fontsize=12)
axes[1].set_title('Power Consumption Range by Hour', fontsize=13, fontweight='bold')
axes[1].set_xticks(range(0, 24))
axes[1].grid(True, alpha=0.3)
axes[1].legend()

plt.tight_layout()
plt.show()

print("\n✓ Hourly pattern analysis complete!")
```

```python
# ════════════════════════════════════════════
# CELL 12: DAILY PATTERN ANALYSIS
# ════════════════════════════════════════════
print("="*60)
print("DAILY PATTERN ANALYSIS")
print("="*60)

# Extract day of week
df_cleaned['DayOfWeek'] = df_cleaned['DateTime'].dt.day_name()
df_cleaned['DayNum'] = df_cleaned['DateTime'].dt.dayofweek

# Calculate daily averages
daily_avg = df_cleaned.groupby(['DayNum', 'DayOfWeek'])['Global_active_power'].mean().reset_index()
daily_avg = daily_avg.sort_values('DayNum')

print("\nAverage Power by Day of Week:")
print(daily_avg[['DayOfWeek', 'Global_active_power']].to_string(index=False))

# Plot
plt.figure(figsize=(12, 6))
plt.bar(daily_avg['DayOfWeek'], daily_avg['Global_active_power'],
        color='coral', edgecolor='black', alpha=0.7)
plt.xlabel('Day of Week', fontsize=12)
plt.ylabel('Average Power (kW)', fontsize=12)
plt.title('Average Power Consumption by Day of Week', fontsize=13, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()

print("\n✓ Daily pattern analysis complete!")
```

```python
# ════════════════════════════════════════════
# CELL 13: SUMMARY & CONCLUSIONS
# ════════════════════════════════════════════
print("="*60)
print("FINAL SUMMARY & CONCLUSIONS")
print("="*60)

print(f"""
📊 DATASET OVERVIEW:
   • Total observations: {df_cleaned.shape[0]:,}
   • Variables analyzed: {df_cleaned.shape[1]}
   • Time period: {df_cleaned['DateTime'].min().strftime('%Y-%m-%d')} to {df_cleaned['DateTime'].max().strftime('%Y-%m-%d')}
   • Duration: {(df_cleaned['DateTime'].max() - df_cleaned['DateTime'].min()).days:,} days ({(df_cleaned['DateTime'].max() - df_cleaned['DateTime'].min()).days/365:.1f} years)

⚡ POWER CONSUMPTION STATISTICS:
   • Mean: {df_cleaned['Global_active_power'].mean():.3f} kW
   • Median: {df_cleaned['Global_active_power'].median():.3f} kW
   • Std Dev: {df_cleaned['Global_active_power'].std():.3f} kW
   • Min: {df_cleaned['Global_active_power'].min():.3f} kW
   • Max: {df_cleaned['Global_active_power'].max():.3f} kW
   • Range: {df_cleaned['Global_active_power'].max() - df_cleaned['Global_active_power'].min():.3f} kW

🔧 DATA QUALITY:
   • Original rows: {df.shape[0]:,}
   • Cleaned rows: {df_cleaned.shape[0]:,}
   • Removed: {df.shape[0] - df_cleaned.shape[0]:,} ({((df.shape[0] - df_cleaned.shape[0])/df.shape[0]*100):.2f}%)
   • Data completeness: {100 - ((df.shape[0] - df_cleaned.shape[0])/df.shape[0]*100):.2f}%

📈 KEY FINDINGS:
""")

# Top correlations
print("   Correlations with Global Active Power:")
correlations = corr_matrix['Global_active_power'].drop('Global_active_power').sort_values(ascending=False)
for var, corr_val in correlations.items():
    print(f"     • {var}: {corr_val:.3f}")

# Peak hours
peak_hour = hourly_avg['mean'].idxmax()
peak_power = hourly_avg['mean'].max()
low_hour = hourly_avg['mean'].idxmin()
low_power = hourly_avg['mean'].min()

print(f"\n   Time Patterns:")
print(f"     • Peak consumption hour: {peak_hour}:00 ({peak_power:.3f} kW)")
print(f"     • Lowest consumption hour: {low_hour}:00 ({low_power:.3f} kW)")

# Day of week patterns
max_day = daily_avg.loc[daily_avg['Global_active_power'].idxmax(), 'DayOfWeek']
min_day = daily_avg.loc[daily_avg['Global_active_power'].idxmin(), 'DayOfWeek']

print(f"     • Highest consumption day: {max_day}")
print(f"     • Lowest consumption day: {min_day}")

print("\n" + "="*60)
print("✓ EXPLORATORY DATA ANALYSIS COMPLETE!")
print("="*60)


```
